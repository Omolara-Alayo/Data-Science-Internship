```python
# Merging the datasets
```

```python
Merged_df_A = pd.merge(Cab_df, Transaction_ID_df, on = 'Transaction ID', how = 'outer')

#Checking the merged_df_A dataframe's shape
print("(rows, columns)",Merged_df_A.shape,"\n")
Merged_df_A.head()
```

```python
Merged_df_B = pd.merge(Customer_ID_df, Merged_df_A, on = 'Customer ID', how = 'outer')

#Checking the merged_df_B dataframe's shape
print("(rows, columns)",Merged_df_B.shape,"\n")
Merged_df_B.head()
```

```python
Merged_df_C = pd.merge(City_df, Merged_df_B, on = 'City', how = 'left')

#Checking the merged_df_B dataframe's shape
print("(rows, columns)",Merged_df_C.shape,"\n")
Merged_df = Merged_df_C
Merged_df.head()
```

```python
#Checking the merged dataframes 
print(Merged_df['Transaction ID'].unique().shape)
Merged_df.dtypes
```

```python
# Converting Population and User columns from object to float for easy analysis

Columns = ['Population','Users']

for item in Columns:
    Merged_df[item] = Merged_df[item].astype(str).str.replace(',', '').astype(float)

# Recheck data types
print(Merged_df.dtypes)
```

```python
# Computing the profit for the two cab Companies

Merged_df['Profit'] = Merged_df['Price Charged'] - Merged_df['Cost of Trip']
Merged_df.head()
```

```python
# Convery Year to integer 

# Replace non-finite values (NaN or inf) with 0
Merged_df['Year'] = Merged_df['Year'].replace([np.nan, np.inf, -np.inf], 0)

# Convert 'Year' column to integer
Merged_df['Year'] = Merged_df['Year'].astype(int)

# Print the result
print(Merged_df['Year'].head())
```

```python
Merged_df.head()
```

```python
#Creating other faetures needed

# Ctrating Profit rate
Merged_df['Profit Rate'] = ((Merged_df['Price Charged'] - Merged_df['Cost of Trip']) / Merged_df['Cost of Trip'])*100

# Creating Profit per Km
Merged_df['Profit per KM'] = Merged_df['Profit']/ Merged_df['KM Travelled']

# Creating the Users ration
Merged_df['User Ratio'] = Merged_df['Users'] / Merged_df['Population']*100
Merged_df.info()
Merged_df.head(2)
```

```python

# Checking the profit distribution for both companies
# Create subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(8, 12))
fig.suptitle('Comparison of Yellow and Pink Companies')

# Profits distributions
sns.histplot(data=Merged_df, x='Profit', kde=True, hue="Company", ax=axes[0])
sns.histplot(data=Merged_df, x='Profit Rate', kde=True, hue="Company", ax=axes[1])
sns.histplot(data=Merged_df, x='Profit per KM', kde=True, hue="Company", ax=axes[2])

# Showing the plot
plt.tight_layout()
plt.show()
```

```python
# Checking neach company yearly profit

def yearly_profit():
    
    # Zip together the 3 columns and create DataFrame
    company_profit = zip(Merged_df['Year'], Merged_df['Profit'], Merged_df['Company'])
    yearly_profit_df = pd.DataFrame(company_profit, columns=["Year", "Profit", "Company"])
    
    # Group by 'Year' and sum up profits
    yearly_profit_sum_df = yearly_profit_df.groupby('Year')['Profit'].sum().round().reset_index()
    
    print(yearly_profit_sum_df)
    

yearly_profit()
```

```python
# Visualizing yealy profit for each comapny


# Grouping by 'Year' and 'Company', sum the profits, and unstack the result
yearly_profit_df = Merged_df.groupby(['Year', 'Company'])['Profit'].sum().unstack()

# Plotting
fig_profit = yearly_profit_df.plot(kind='bar', figsize=(16, 4))
fig_profit.set_title('Companies Yearly Profit Distribution')
fig_profit.set_ylabel('Profits')

plt.show()
```

```python
# Visualizing the profit made by these comapnies in relation to different cities

# Grouping by 'City' and 'Company', sum the profits, and unstack the result
Cprofit_df = Merged_df.groupby(['City', 'Company'])['Profit'].sum().unstack()

# Plotting
fig_CP = Cprofit_df.plot(kind='bar', figsize=(16, 4))
fig_CP.set_title('Profits made from different Cities')
fig_CP.set_ylabel('Profits')

plt.show()
```

```python
# Profit trend across the three years

fig=Merged_df.groupby(['Year','Company']).sum().unstack().plot(kind='line',y='Profit', stacked=False,figsize=(18,5))
fig.set_title('Profits for different years')
fig.set_ylabel('Profits')
```

```python
# Comparing each cab company's profit rate against the total cab market


Yearly_Profit_Rate = Merged_df.groupby(['Year', 'Company']).Profit.sum().unstack() / \
                     Merged_df.groupby(['Year', 'Company'])['Cost of Trip'].sum().unstack() * 100

Cab_Annual_PR = Merged_df.groupby(['Year']).Profit.sum() / Merged_df.groupby(['Year'])['Cost of Trip'].sum() * 100
Yearly_Profit_Rate['Cab Market'] = Cab_Annual_PR

print("Yearly Profit Rate of Cab Market:")
print(Yearly_Profit_Rate)

# Plotting
ax = Yearly_Profit_Rate.plot(kind='bar', stacked=False, title='Yearly Profit Rate Comparison')
plt.show()
```

```python

# Checking the market share of both cab companies 
# Counting transactions per city and company
Cab_transaction = Merged_df.groupby(['City', 'Company']).size().unstack()

# Adding row for total transactions across all cities
Cab_transaction.loc['All Cities'] = Cab_transaction.sum()

# Calculating total transactions and market shares
Cab_transaction['City_Market'] = Cab_transaction['Pink Cab'] + Cab_transaction['Yellow Cab']
Cab_transaction['Pink_Market_Share'] = Cab_transaction['Pink Cab'] / Cab_transaction['City_Market']
Cab_transaction['Yellow_Market_Share'] = Cab_transaction['Yellow Cab'] / Cab_transaction['City_Market']

# Plotting market share comparison
ax = Cab_transaction[['Pink_Market_Share', 'Yellow_Market_Share']].plot(kind='barh', stacked=False, title='Market Share Comparison across Cities')  
plt.xlabel('Market Share')
plt.ylabel('City')
plt.show()
```
